```python
with open(r'E:\PY\wt_stats_parser\vehicles_rus.json', encoding='UTF-8') as file:
    vehicles_rus = json.load(file)
```

```python
RB_TO_BR = {
    0: 0.7, 1: 1.0, 2: 1.3, 3: 1.7, 4: 2.0, 5: 2.3, 6: 2.7, 7: 3.0,
    8: 3.3, 9: 3.7, 10: 4.0, 11: 4.3, 12: 4.7, 13: 5.0, 14: 5.3,
    15: 6.0, 16: 6.3, 17: 6.7, 18: 7.0, 19: 7.3, 20: 7.7, 21: 8.0,
    22: 8.3, 23: 8.7, 24: 9.0, 25: 9.3, 26: 9.7, 27: 10.0,
    28: 10.3, 29: 10.7, 30: 11.0, 31: 11.3, 32: 11.7, 33: 12.0,
    34: 12.3, 35: 12.7
}
```

```python
def get_vehicle_info(vehicle_name, vehicles_db):
    """
    Ищет в базе vehicles_db технику по отображаемому имени.
    Возвращает: тип на русском, BR (в режиме RB)
    """
    for item in vehicles_db:
        if len(item) < 8:
            continue  # защита от битых записей
        display_name = item[1]
        br_dict = item[4]       # {'ab': 31, 'rb': 31, 'sb': 31}
        type_data = item[7]     # [[type_code, type_rus, color], ...]

        if display_name == vehicle_name:
            br_rb = br_dict['rb']  # например, 31
            real_br = RB_TO_BR.get(br_rb, None)  # конвертируем в 10.3

            vehicle_type_rus = "Неизвестно"
            if type_data and len(type_data) > 0 and len(type_data[0]) > 1:
                vehicle_type_rus = type_data[0][1]  # например, "Тяжёлый танк"

            return {
                'type': vehicle_type_rus,
                'br': real_br,
                'country': item[2].title(),  # China → China
                'rank': item[3]             # ранг (не BR)
            }
    return None
```

```python
veh = 'Me 262 A-1a/U1'
```

```python
info = get_vehicle_info(veh, vehicles_rus)
if info:
    print(f"Тип: {info['type']}")
    print(f"BR (RB): {info['br']}")
    print(f"Страна: {info['country']}")
else:
    print("Техника не найдена")
```
