```python
import requests
import json
from urllib.parse import urlencode
from bs4 import BeautifulSoup
import re
```

```python
#### готовые функции
```

```python
def fetch_wt_units_from_url(url):
    """
    Загружает страницу по URL и извлекает JSON из window.WT_UnitList.
    Возвращает список техники или None, если не удалось.
    """
    print(f"🌐 Загружаем: {url}")
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"❌ Ошибка загрузки {url}: {e}")
        return None

    html_content = response.text

    # Паттерн: ищем window.WT_UnitList = 'JSON_STRING'
    pattern = r'window\.WT_UnitList\s*=\s*\'([^\']+)\''
    match = re.search(pattern, html_content)

    if match:
        json_string = match.group(1)
        try:
            units = json.loads(json_string)
            print(f"✅ Найдено {len(units)} записей в {url}")
            return units
        except json.JSONDecodeError as e:
            print(f"❌ Ошибка декодирования JSON в {url}: {e}")
            return None
    else:
        print(f"❌ Не найден window.WT_UnitList в {url}")
        return None

def combine_vehicle_data(urls):
    """
    Скачивает данные с нескольких URL и объединяет их.
    """
    all_units = []

    for url in urls:
        units = fetch_wt_units_from_url(url)
        if units:
            all_units.extend(units)

    print(f"\n✅ Всего техники собрано: {len(all_units)}")
    return all_units

# === ЗАПУСК ===
if __name__ == "__main__":
    urls = [
        'https://wiki-ru.warthunder.com/ground?from=en_lang&v=l',
        'https://wiki-ru.warthunder.com/aviation?v=l',
        'https://wiki-ru.warthunder.com/helicopters?v=l'
    ]

    combined_data = combine_vehicle_data(urls)

    if combined_data:
        output_file = 'vehicles_rus_upd.json'
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(combined_data, f, ensure_ascii=False, indent=2)
        print(f"\n💾 Все данные сохранены в: {output_file}")
    else:
        print("\n❌ Не удалось получить данные ни с одной страницы.")
```
